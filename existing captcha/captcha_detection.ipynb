```python
# Import libraries
from matplotlib import pyplot as plt
import numpy as np
import random
 
#num = random.random(1,50,4)
#print(num)

 
# Function to generate
# and append them
# start = starting range,
# end = ending range
# num = number of
# elements needs to be appended

max_value = 0
max_index = 0

def Rand(start, end, num):
    res = []
 
    for j in range(num):
        res.append(random.randint(start, end))
    
 
    return res
 
# Driver Code
num = 3
start = 1
end = 100

# Creating dataset
cars = [
        'a', 'b', 'c']
 
data = Rand(start, end, num)

max_value = max(data)
max_index = data.index(max_value)
#print(max_value)
#print(max_index)
 
# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = cars)
 
# show plot
plt.show()
letter = input("Enter the letter which has max contribution : ")
key = 0
if(letter == 'a'):
    key = 0
if(letter == 'b'):
    key = 1
if(letter == 'c'):
    key = 2
if(letter==cars[max_index] or max_value == data[key]):
    print("Success")
else:
    print("Fail")
```
